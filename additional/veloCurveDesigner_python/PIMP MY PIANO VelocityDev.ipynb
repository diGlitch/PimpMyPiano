## Settings and Methods

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
MAX_DURATION = 300
MIN_DURATION = 3
```

```python
# Map a value from one range (f and t) into another one (maxLim and minLim)
def map(x, f, t, minLim, maxLim):
    return (t + (f - t) * ((x - minLim) / (maxLim - minLim)));
```

```python
# print given array in the folling format: { x_0, x_1, x_2, x_3, x_4 };
def printAsCArray(h):
    string = "{"
    for i in h:
        string = string + str(int(np.round(i))) + ", "
    string = string[:-2] + "};"
    print(string)
```

```python
# tranform a given duration time x to a velocity value between 0 and 127
def getVelo(x):
    return map(x, 1, 127, MIN_DURATION, MAX_DURATION)
```

## Linear velocity

```python
times = np.arange(MIN_DURATION, MAX_DURATION+1)
velo1 = getVelo(times)

df = pd.Series(velo1, index=times)
plt.title("Velo vs key press duration")
plt.xlabel("Duration in ms")
plt.ylabel("Velocity")
df.plot()

printAsCArray(velo1)
```

## Non-Linear Velocity

### velo2
#### More weight on higher duration to make silent play easier


```python
timesNorm = map(times, 0, 1, MIN_DURATION, MAX_DURATION)
```

```python
def funcVelo2(x):
    return (x * x * x) * (1)

def funcExp(x):
    return np.exp(x*(4))

newTimes = funcVelo2(timesNorm)

velo2 = map(newTimes, 0, 127, np.max(newTimes), np.min(newTimes))

df = pd.Series(velo2, index=times)
plt.title("Velo vs key press duration")
plt.xlabel("Duration in ms")
plt.ylabel("Velocity")
df.plot()

printAsCArray(velo2)
```

```python
a = 0
b = 1
c = 0
d = 1

def funcVelo4(x):
    c1 = np.multiply(np.power(1-x,3), a)
    c2 = np.multiply(np.multiply(np.multiply(np.power(1-x,2), 3), b), x)
    c3 = np.multiply(np.multiply(np.multiply(1-x, np.power(x,2)), c), 3)
    c4 = np.multiply(np.power(x,3), d)
    return c1 + c2 + c3 + c4

newTimes = funcVelo4(timesNorm)

velo4 = map(newTimes, 0, 127, np.max(newTimes), np.min(newTimes))

df = pd.Series(velo4, index=times)
plt.title("Velo vs key press duration")
plt.xlabel("Duration in ms")
plt.ylabel("Velocity")
df.plot()

printAsCArray(velo4)
```
